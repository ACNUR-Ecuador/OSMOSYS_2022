```python
import os
import pandas as pd
import json
from requests.exceptions import HTTPError
import activityinfo
import osmosys.osmosys
import osmosys.Backups
```

```python
month = 'JUNIO'
month_number = '2022-06'
year = 2022
test = False
```

```python
folder = str(year) + '-' + month
path = os.path.join('.', folder)
```

```python

dir_path = path
backupFiles = []
for path in os.scandir(dir_path):
    if path.is_file():
        if (path.name.startswith('new_ids')):
            backupFiles.append(path.name)

```

```python
dfBK = None
for file in backupFiles:
    print(file)
    df = pd.read_csv(os.path.join(dir_path, file))
    if dfBK is None:
        dfBK = df
    else:
        if df.shape[0] > 0:
            dfBK = pd.concat([dfBK, df], sort=False, ignore_index=True)
```

```python
print(dfBK)
```

```python
changes = []
for index, record in dfBK.iterrows():
    change = {
        "formId": record['formId'],
        "recordId": record['recordId'],
        "parentRecordId": None,
        "deleted": True,
        "fields": None
    }
    changes.append(change)
```

```python
print('records to delete: ' + str(len(changes)))
print(changes)
```

```python
finalJson = json.dumps(changes)
try:
    os.mkdir(folder)
except Exception:
    pass
print(folder)
text_file = open(os.path.join(folder, 'rollback' + ".json"), "w")
# write string to file
n = text_file.write(finalJson)
# close file
text_file.close()
```

```python
if (test):
    print(
        '--------------------------------------' + 'Rolll back' + '------------------------------------------------------')

else:

    if len(changes) > 0:
        print('se envia a AI')
        for change in changes:
            totalChanges = {
                'changes': [change]
            }
            finalJson = json.dumps(totalChanges)
            print(finalJson)
            print()
            print()
            try:
                client = activityinfo.Client(token=osmosys.osmosys.getToken(),
                                             base_url='https://www.activityinfo.org/resources')
                client.post_resource(path='update', body=finalJson)
            except HTTPError as e:
                code = e.response.status_code
                print('error:')
                print(code)

                print(e)
    else:
        print('nada que enviar a AI')

print(
    '--------------------------------------' + 'Rolll back' + '------------------------------------------------------')
```
